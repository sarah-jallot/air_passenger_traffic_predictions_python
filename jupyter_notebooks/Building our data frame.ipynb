```python
import pandas as pd
```

```python
external_data = pd.read_csv('external_data_morning.csv')
external_data.head()
```

```python
external_data.columns
```

```python
# Changing the names of our columns to make them easier to understand

columns_rename = [w.replace('d_', 'departure_') for w in external_data]
columns_rename = [w.replace('a_', 'arrival_') for w in columns_rename]

external_data.columns = columns_rename
external_data.head()
```

```python
# Adding a column with the month and another one with the year

external_data.loc[:,"Date"] = pd.to_datetime(external_data.loc[:,"Date"])
external_data['year'] = external_data.loc[:,'Date'].dt.year
external_data['month'] = external_data.loc[:,'Date'].dt.month
external_data.head()
```

```python
# Importing our performance by month and year for each airport

performance = pd.read_csv("airport_performanceClean.csv")
performance.head()
```

```python
# Merging our 2 dataframes 

external_data = pd.merge(external_data, performance, left_on=(['year', 'month', 'AirPort']), right_on=(['year', 'month', 'AirPort']), how='left')
external_data.head()
```

```python
airport_list = external_data['AirPort'].unique()
airport_list
```

```python
# Importing the number of passengers in 2011 by airport by month 

passengers_2011 = pd.read_csv('2011.csv',index_col=False)
passengers_2012 = pd.read_csv('2012.csv',index_col=False)
passengers_2013 = pd.read_csv('2013.csv',index_col=False)
```

```python
passengers_2011
```

```python
passengers_2011['ORIGIN'].unique()
```

```python
passengers_2011['DEST'].unique()
```

```python
passengers_2011.columns
```

```python
# Dropping a column of just NaN

passengers_2011 = passengers_2011.drop('Unnamed: 4', axis=1)
passengers_2012 = passengers_2012.drop('Unnamed: 4', axis=1)
passengers_2013 = passengers_2013.drop('Unnamed: 4', axis=1)
```

```python
# Dropping rows with missing values 

passengers_2011 = passengers_2011.dropna()
passengers_2012 = passengers_2012.dropna()
passengers_2013 = passengers_2013.dropna()
```

```python
# adding the columns with the year 

passengers_2011['year'] = 2011
passengers_2012['year'] = 2012
passengers_2013['year'] = 2013
```

```python
# Keeping only the connexions that we need 

passengers_2011 = passengers_2011[passengers_2011['ORIGIN'].isin(airport_list)]
passengers_2011 = passengers_2011[passengers_2011['DEST'].isin(airport_list)]
passengers_2011 = passengers_2011[passengers_2011['ORIGIN'] != passengers_2011['DEST']]

passengers_2012 = passengers_2012[passengers_2012['ORIGIN'].isin(airport_list)]
passengers_2012 = passengers_2012[passengers_2012['DEST'].isin(airport_list)]
passengers_2012 = passengers_2012[passengers_2012['ORIGIN'] != passengers_2012['DEST']]

passengers_2013 = passengers_2013[passengers_2013['ORIGIN'].isin(airport_list)]
passengers_2013 = passengers_2013[passengers_2013['DEST'].isin(airport_list)]
passengers_2013 = passengers_2013[passengers_2013['ORIGIN'] != passengers_2013['DEST']]
```

```python
passengers_2011
```

```python
# Set the good index column

passengers_2011 = passengers_2011.set_index('PASSENGERS')
passengers_2012 = passengers_2012.set_index('PASSENGERS')
passengers_2013 = passengers_2013.set_index('PASSENGERS')
```

```python
# We can see that there are way to many values and thus look at duplicated rows

passengers_2011.duplicated().sum()
```

```python
# Removing duplicates (we want at max 20*19*12=4560 rows)

passengers_2011 = passengers_2011.drop_duplicates()
passengers_2012 = passengers_2012.drop_duplicates()
passengers_2013 = passengers_2013.drop_duplicates()
```

```python
passengers = pd.concat([passengers_2011, passengers_2012, passengers_2013])
passengers.reset_index(level=0, inplace=True)
passengers
```

```python
# Merging this into our original data frame 

external_data = pd.merge(external_data, passengers, left_on=(['year', 'month', 'AirPort']), right_on=(['year', 'MONTH', 'ORIGIN']), how='left')
```

```python
external_data.head()
```

```python
external_data = external_data.drop('ORIGIN', axis=1)
```

```python
external_data
```

```python
# Deliting rows when departure = arrival
external_data = external_data[external_data['AirPort'] != external_data['DEST']]
```

```python
external_data
```

```python
external_data = external_data.drop(['year', 'month', 'MONTH'], axis=1)
```

```python
external_data.head()
```

```python
# Rename our columns 
external_data = external_data.rename(columns={"PASSENGERS": "average_passengers", "DEST": "Destination"})
```

```python
external_data.head()
```

```python
# Building our CSV file

external_data_csv = external_data.to_csv('external_data.csv', index = None, header=True) 
#Don't forget to add '.csv' at the end of the path
print(external_data)
```
