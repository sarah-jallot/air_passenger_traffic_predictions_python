```python
import pandas as pd
```

```python
data_vic = pd.read_csv('external_data_vic.csv')
```

```python
data_vic.head()
```

```python
data_sarah = pd.read_csv('external_data_sarah.csv')
```

```python
data_sarah.head()
```

```python
data_sarah = data_sarah[['Date', 'AirPort', 'HigherTraffic_Dates', 'M', 'N', 'S', 'W']]
data_sarah.head()
```

```python
data = pd.merge(data_vic, data_sarah, how='left', left_on=['Date', 'AirPort'], right_on=['Date', 'AirPort'])
data.head()
```

```python
data_csv = data.to_csv('external_data.csv', index = None, header=True) 
#Don't forget to add '.csv' at the end of the path
print(data)
```
