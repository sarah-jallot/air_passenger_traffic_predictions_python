```python
# Adding regular weekends to the df

data['Date'] = pd.to_datetime(data['Date'])

# Monday = 0 on Python
data['Weekday'] = data['Date'].dt.weekday
data.head()

# Building a column with weekends
data['Weekend'] = data['Weekday'].apply(lambda x: '1' if x==5 or x==6 else '0')

# We consider that people take planes on friday and sunday when traveling for the weekend 
data['Weekend_approx'] = data['Weekday'].apply(lambda x: '1' if x==4 or x==6 else '0')

data.head()
```

```python
# We want to add hubs as special airports 

data['Busiest_top5'] = data['AirPort'].apply(lambda x: '1' if x in ['ATL', 'LAX', 'ORD', 'DFW', 'DEN'] else '0')
data.head()
```

```python
# Building a column that outputs if the day is a holiday or not in the state of the airport

import numpy as np
State_holidays = []

for i in np.arange(len(data)):
    if data['Date'][i] in holidays.CountryHoliday('US', prov=None, state=data['State'][i]):
        State_holidays.append('1')
    else:
        State_holidays.append('0')
```

```python
# Putting this column into our dataframe 
data['State_holidays'] = State_holidays
```
